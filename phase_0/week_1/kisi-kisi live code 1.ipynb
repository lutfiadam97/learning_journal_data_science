Hitung luas jika dual_sim = 1, jika tidak maka hitung kurang:

- luas = tinggi(px_height)*lebar(px_width)
- kurang = tinggi - lebar
- bikin kolom baru yang isinya dari hasil perhitungan luas dan kurang

```python
def luas(tinggi,lebar):
    hasil = tinggi * lebar
    return hasil

def kurang(tinggi,lebar):
    hasil_k = tinggi- lebar
    return hasil_k

```

```python
df_new = df.copy()

lk = [] # untuk masukan hasil kedalam list
j = 0 #index baris
#df_new['dual_sim'] untuk nilai kondisi awal

for i in df_new['dual_sim']:
    if i == 1:
        lk.append(luas(df_new['px_height'][j], df_new['px_width'][j]))
        
    else:
        lk.append(kurang(df_new['px_height'][j], df_new['px_width'][j]))
    j+=1
    #print(df_new['px_height'][j]) #pengecekan alur nilai

#bikin kolom
df_new['l/k'] = lk
```
