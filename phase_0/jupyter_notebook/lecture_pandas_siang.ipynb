```python
# pake query -> lebih ringkas
df.query("team_id == 'BLB' and pts > 100 and opp_pts > 100 and year_id == 1953") 
```

### Grouping and Aggregating Data

```python
# kelompokkin fran_id terus hitung nilai sesuai dgn pts
df.groupby('fran_id')['pts'].sum()
```

```python
df[
    (df['year_id'] > 2010) &
    (df['fran_id'] == "Knicks")
].groupby("year_id")['pts'].sum()
```

```python
df[
    (df['year_id'] > 2010) &
    (df['fran_id'] == "Knicks")
].groupby(["year_id", "game_location"])['pts'].sum()
```
